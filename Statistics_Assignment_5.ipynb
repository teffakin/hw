The hourly wages in a particular industry are normally distributed with mean 13.20 Dollar and standard deviation 2.50 Dollar. A company in this industry employs 40 workers, paying them an average of $12.20 per hour. Can this company be accused of paying substandard wages? Use an Î± = .01 level test.

```python
mu = 13.20
sigma = 2.50

n=40

#H0 = 12.20 
#H1 < 12.20 (is significantly lower than 13.20)

alpha = 0.01
```

```python
ttest = (12.20-13.20)/(2.50/np.sqrt(40))
ttest
```

```python
p_value = stats.norm.cdf(ttest)
p_value
```

```python
if p_value < alpha :
    print("Reject the null")
else :
    print("Fail to reject the null")
```
